```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
class Portfolio:
    def __init__(self):
        self.data = pd.DataFrame(columns=["Asset", "Type", "Units", "Price", "Dividend Yield"])
```

```python
def add_asset(self, name, asset_type, units, price, dividend_yield):
        new_asset = {
            "Asset": name,
            "Type": asset_type,
            "Units": units,
            "Price": price,
            "Dividend Yield": dividend_yield
        }
        self.data = pd.concat([self.data, pd.DataFrame([new_asset])], ignore_index=True)
        print(f"Added {units} units of {name} to the portfolio.")
```

```python
def update_price(self, name, new_price):
        if name in self.data["Asset"].values:
            self.data.loc[self.data["Asset"] == name, "Price"] = new_price
            print(f"Updated price of {name} to {new_price}.")
        else:
            print(f"Asset {name} not found in the portfolio.")
```

```python
  def generate_report(self):
        self.data["Total Value"] = self.data["Units"] * self.data["Price"]
        self.data["Annual Dividend Income"] = (self.data["Dividend Yield"] / 100) * self.data["Total Value"]
        total_portfolio_value = self.data["Total Value"].sum()
        total_dividend_income = self.data["Annual Dividend Income"].sum()

        print("\nPortfolio Summary:")
        print(self.data.to_string(index=False))
        print(f"\nTotal Portfolio Value: ₹{total_portfolio_value:.2f}")
        print(f"Total Annual Dividend Income: ₹{total_dividend_income:.2f}")
```

```python
   def adjust_portfolio_value(self, target_value):
        current_value = (self.data["Units"] * self.data["Price"]).sum()
        scaling_factor = target_value / current_value
        self.data["Price"] = self.data["Price"] * scaling_factor
        print(f"\nAdjusted portfolio to a total value of ₹{target_value}.")
```

```python
def simulate_random_walk_growth(self, start_value, end_value, days):
        portfolio_values = [start_value]
        for _ in range(1, days):
            daily_change = np.random.uniform(-0.02, 0.03)  # -2% to +3% change each day
            new_value = portfolio_values[-1] * (1 + daily_change)
            portfolio_values.append(new_value)

        portfolio_values = np.array(portfolio_values)
        scaling_factor = end_value / portfolio_values[-1]
        portfolio_values *= scaling_factor

        return portfolio_values

```

```python
 def simulate_nifty(self, start_value, max_value, min_value, days):
        nifty_values = [start_value]
        rising = True

        for _ in range(1, days):
            if rising:
                daily_change = np.random.uniform(0, 0.002)  # Small upward movement
                new_value = nifty_values[-1] * (1 + daily_change)
                if new_value >= max_value:
                    new_value = max_value
                    rising = False
            else:
                daily_change = np.random.uniform(-0.003, 0)
                new_value = nifty_values[-1] * (1 + daily_change)
                if new_value <= min_value:
                    new_value = min_value
                    rising = True

            nifty_values.append(new_value)

        return nifty_values
```

```python
  def compare_performance(self, benchmark_data):
        portfolio_values = self.simulate_random_walk_growth(21000, 26700, len(benchmark_data))

        plt.figure(figsize=(10, 6))
        plt.plot(benchmark_data["Date"], benchmark_data["Index"], label="NIFTY 50 Value", color="blue")
        plt.plot(benchmark_data["Date"], portfolio_values, label="Portfolio Value", color="green")
        plt.xlabel("Date")
        plt.ylabel("Value (₹)")
        plt.title("Portfolio vs NIFTY 50 Value Comparison")
        plt.legend()
        plt.grid(True, alpha=0.5)
        plt.tight_layout()
        plt.show()
```

```python
portfolio = Portfolio()
```

```python
portfolio.add_asset("IRCTC", "Stock", 50, 700, 0.5)
portfolio.add_asset("Zomato", "Stock", 100, 75, 0.0)
portfolio.add_asset("Adani", "Stock", 10, 2000, 0.4)
portfolio.add_asset("Reliance", "Stock", 20, 2300, 0.8)
portfolio.add_asset("Tata", "Stock", 30, 900, 0.6)
portfolio.add_asset("ITC", "Stock", 40, 450, 1.5)
```

```python
portfolio.adjust_portfolio_value(20000)
portfolio.generate_report()
nifty_index = 25200
benchmark_data = pd.DataFrame({
    "Date": pd.date_range(start="2023-01-01", periods=252, freq="B"),
    "Index": portfolio.simulate_nifty(nifty_index, 25100, 25000, 252)
})
portfolio.compare_performance(benchmark_data)

```
